```python
def coarseGrainHbl_MAX(hblset, latsel, lonsel):
    periodlength = 24
    periodnum = int(hblset.dims['time'] / periodlength)
    hblsel = hblset['KPPhbl'].sel(lat=latsel,lon=lonsel,method='nearest')
    perioddepths = np.empty(periodnum)
    hblcoarsegrain = np.empty(0)
    for index in range(0,periodnum):
        perioddepths[index] = hblsel[index*periodlength:(index+1)*periodlength].max(dim='time').values
        hblcoarsegrain = np.concatenate((hblcoarsegrain, np.full(periodlength, perioddepths[index])),casting='same_kind')
    hMax = xr.DataArray(data = hblcoarsegrain, dims = hblsel.dims, coords = hblsel.coords)
    return hMax
```

```python
def coarseGrainHbl_SUNRISE(hblset, latsel, lonsel):
    periodlength = 24
    periodnum = int(hblset.dims['time'] / periodlength)
    hblsel = hblset['KPPhbl'].sel(lat=latsel,lon=lonsel,method='nearest')
    perioddepths = np.empty(periodnum)
    hblcoarsegrain = np.empty(0)
    sunrisetime = 10 #NOTE IN A FINAL VERSION OF THIS CODE, SUNRISE TIME WILL BE A FUNCTION OF LONGITUDE
    for index in range(0,periodnum):
        perioddepths[index] = hblsel[index*periodlength + sunrisetime].values
        hblcoarsegrain = np.concatenate((hblcoarsegrain, np.full(periodlength, perioddepths[index])),casting='same_kind')
    hSUNRISE = xr.DataArray(data = hblcoarsegrain, dims = hblsel.dims, coords = hblsel.coords)
    return hSUNRISE
def instantaneousHBL(hblset, latsel, lonsel):
    return hblset['KPPhbl'].sel(lat=latsel,lon=lonsel,method='nearest')
```

```python
def coarseGrainHbl_LPFilter(hblset,latsel,lonsel):
    periodlength = 24
    periodnum = int(hblset.dims['time'] / periodlength)
    hblfilt = dsp.lowpass(dsp.lowpass(dsp.lowpass(hblset['KPPhbl'], 1.0/0.4,dim='lat' ), 1.0/0.4, dim='lon'), 1.0 / 48.0, dim='time')
    hblsel = hblfilt.sel(lat=latsel,lon=lonsel,method='nearest')
    perioddepths = np.empty(periodnum)
    hblcoarsegrain = np.empty(0)
    sunrisetime = 10 #NOTE IN A FINAL VERSION OF THIS CODE, SUNRISE TIME WILL BE A FUNCTION OF LONGITUDE
    for index in range(0,periodnum):
        perioddepths[index] = hblsel[index*periodlength + sunrisetime].values
        hblcoarsegrain = np.concatenate((hblcoarsegrain, np.full(periodlength, perioddepths[index])),casting='same_kind')
    hSUNRISE = xr.DataArray(data = hblcoarsegrain, dims = hblsel.dims, coords = hblsel.coords)
    return hSUNRISE
def coarseGrainHbl_CONST(hblset,latsel,lonsel,zlayer):
    periodlength = 24
    periodnum = int(hblset.dims['time'] / periodlength)
    hblsel = hblset.sel(lat=latsel,lon=lonsel,method='nearest')
    perioddepths = np.empty(periodnum)
    hblcoarsegrain = np.empty(0)
    for index in range(0,periodnum):
        perioddepths[index] = zlayerToDepth(zlayer)
        hblcoarsegrain = np.concatenate((hblcoarsegrain, np.full(periodlength, perioddepths[index])),casting='same_kind')
    hSUNRISE = xr.DataArray(data = hblcoarsegrain, dims = hblsel.dims, coords = hblsel.coords)
    return hSUNRISE
```

```python
def getThetaTendency(hblcoarsegrain, Theta, latsel, lonsel):
    hblcg = hblcoarsegrain.values #hblcoarse grain is a dataarray and getThickWeightedVector takes dataarrays
    weightVector = getThickWeightedVector(hblcg)
    wv = np.swapaxes(weightVector, 0, 1)
    Thetasel = Theta['Theta'].sel(lat=latsel,lon=lonsel,method='nearest')
    wpdf = xr.DataArray(data = wv, dims = Thetasel.dims, coords = Thetasel.coords)
    weightedThetaSel = (Thetasel[0:wpdf.shape[0],:]*wpdf).sum(dim='Zlayers')
    ThetaTendency = weightedThetaSel.diff(dim='time') / (3600)
    return ThetaTendency

def getThetaTendencyEstimate(hblcoarsegrain, HAdv, Theta, Zeta, Q, latsel, lonsel):
    hblcg = hblcoarsegrain.values #hblcoarse grain is a dataarray and getThickWeightedVector takes dataarrays
    weightVector = getThickWeightedVector(hblcg)
    wv = np.swapaxes(weightVector, 0, 1)
    Thetasel = Theta['Theta'].sel(lat=latsel,lon=lonsel,method='nearest')
    zetasel = Zeta['Zeta'].sel(lat=latsel,lon=lonsel,method='nearest')
    qsel = Q['oceQnet'].sel(lat=latsel,lon=lonsel,method='nearest')
    HAdvsel = HAdv['HAdv'].sel(lat=latsel,lon=lonsel,method='nearest')
    wpdf = xr.DataArray(data = wv, dims = Thetasel.dims, coords = Thetasel.coords)
    weightedThetaSel = (Thetasel[0:wpdf.shape[0],:]*wpdf).sum(dim='Zlayers')
    weightedHAdv = (HAdvsel[0:wpdf.shape[0],:]*wpdf).sum(dim='Zlayers')
    ekman = zetasel / hblcoarsegrain * weightedThetaSel
    heatflux = qsel / (rhoConst * Cp)
    Estimate = weightedHAdv*-1 + ekman + heatflux
    return Estimate
def getHAdvEstimate(hblcoarsegrain, HAdv, Theta, Zeta, Q, latsel, lonsel):
    hblcg = hblcoarsegrain.values #hblcoarse grain is a dataarray and getThickWeightedVector takes dataarrays
    weightVector = getThickWeightedVector(hblcg)
    wv = np.swapaxes(weightVector, 0, 1)
    Thetasel = Theta['Theta'].sel(lat=latsel,lon=lonsel,method='nearest')
    zetasel = Zeta['Zeta'].sel(lat=latsel,lon=lonsel,method='nearest')
    qsel = Q['oceQnet'].sel(lat=latsel,lon=lonsel,method='nearest')
    HAdvsel = HAdv['HAdv'].sel(lat=latsel,lon=lonsel,method='nearest')
    wpdf = xr.DataArray(data = wv, dims = Thetasel.dims, coords = Thetasel.coords)
    weightedThetaSel = (Thetasel[0:wpdf.shape[0],:]*wpdf).sum(dim='Zlayers')
    weightedHAdv = (HAdvsel[0:wpdf.shape[0],:]*wpdf).sum(dim='Zlayers')
    ekman = zetasel / hblcoarsegrain * weightedThetaSel
    heatflux = qsel / (rhoConst * Cp)
    return weightedHAdv*-1
def getEkman(hblcoarsegrain, HAdv, Theta, Zeta, Q, latsel, lonsel):
    hblcg = hblcoarsegrain.values #hblcoarse grain is a dataarray and getThickWeightedVector takes dataarrays
    weightVector = getThickWeightedVector(hblcg)
    wv = np.swapaxes(weightVector, 0, 1)
    Thetasel = Theta['Theta'].sel(lat=latsel,lon=lonsel,method='nearest')
    zetasel = Zeta['Zeta'].sel(lat=latsel,lon=lonsel,method='nearest')
    qsel = Q['oceQnet'].sel(lat=latsel,lon=lonsel,method='nearest')
    HAdvsel = HAdv['HAdv'].sel(lat=latsel,lon=lonsel,method='nearest')
    wpdf = xr.DataArray(data = wv, dims = Thetasel.dims, coords = Thetasel.coords)
    weightedThetaSel = (Thetasel[0:wpdf.shape[0],:]*wpdf).sum(dim='Zlayers')
    weightedHAdv = (HAdvsel[0:wpdf.shape[0],:]*wpdf).sum(dim='Zlayers')
    ekman = zetasel / hblcoarsegrain * weightedThetaSel
    heatflux = qsel / (rhoConst * Cp)
    Estimate = weightedHAdv*-1 + ekman + heatflux
    return ekman

def getHeatFlux(hblcoarsegrain, HAdv, Theta, Zeta, Q, latsel, lonsel):
    hblcg = hblcoarsegrain.values #hblcoarse grain is a dataarray and getThickWeightedVector takes dataarrays
    weightVector = getThickWeightedVector(hblcg)
    wv = np.swapaxes(weightVector, 0, 1)
    Thetasel = Theta['Theta'].sel(lat=latsel,lon=lonsel,method='nearest')
    zetasel = Zeta['Zeta'].sel(lat=latsel,lon=lonsel,method='nearest')
    qsel = Q['oceQnet'].sel(lat=latsel,lon=lonsel,method='nearest')
    HAdvsel = HAdv['HAdv'].sel(lat=latsel,lon=lonsel,method='nearest')
    wpdf = xr.DataArray(data = wv, dims = Thetasel.dims, coords = Thetasel.coords)
    weightedThetaSel = (Thetasel[0:wpdf.shape[0],:]*wpdf).sum(dim='Zlayers')
    weightedHAdv = (HAdvsel[0:wpdf.shape[0],:]*wpdf).sum(dim='Zlayers')
    ekman = zetasel / hblcoarsegrain * weightedThetaSel
    heatflux = qsel / (rhoConst * Cp)
    Estimate = weightedHAdv*-1 + ekman + heatflux
    return heatflux
```

So using calculation of variations, I believe I can come up with the most systematic way to properly determine the best way to determine hbl. Let's write down an "action" that looks like the following. This is also the correct least squares residual formulation I believe.

$$ S = \int dt \int\limits_{x=x_0}^{x=x_0 + w_x} dx \int\limits_{y=y_0}^{y=y_0 + w_y} dy L[H;x,y,t] 
    = \int dt \int\limits_{x=x_0}^{x=x_0 + w_x} dx \int\limits_{y=y_0}^{y=y_0 + w_y} dy \bigg[\frac{1}{H}\int\limits_{z=0}^{z= H(x,y,t)} dz\bigg(\frac{d\theta}{dt} - \frac{d\theta_{\text{est}}}{dt}\bigg)\bigg]^2$$
    
Because $\theta_{\text{est}}$ depends on $H$, an analytical calculation of the optimization calculation is a nontrivial mess of product rule derivatives. Not necessarily difficult, but given that this is a numerical project, we can just calculate the above residual and determine which is the minimum. Numerically.

Note:

$$
\frac{d\theta_{\text{est}}}{dt} = - u \frac{d\theta}{dx} - v \frac{d\theta}{dy} + \frac{w_{\text{ek}}}{H} \theta + \frac{Q_{\text{net}}}{\rho C_p H}$$

```python
# I don't think any of the code in this cell is necessary any longer, but I'll save it in the case that I am wrong. (9/20 1:11 pm ET)
def getThetaTendencyAtLayer(Zsel, Theta, latsel, lonsel):
    #hblcg = hblcoarsegrain.values #hblcoarse grain is a dataarray and getThickWeightedVector takes dataarrays
    #weightVector = getThickWeightedVector(hblcg)
    #wv = np.swapaxes(weightVector, 0, 1)
    Thetasel = Theta['Theta'].sel(lat=latsel,lon=lonsel,method='nearest')
    ThetaTendencyAtLayer = Thetasel[zsel].diff(dim='time') / (3600)
    return ThetaTendencyAtLayer

def getThetaTendencyEstimateAtLayer(Zsel, HAdv, Theta, Zeta, Q, latsel, lonsel):
    Thetasel = Theta['Theta'].sel(lat=latsel,lon=lonsel,method='nearest')
    zetasel = Zeta['Zeta'].sel(lat=latsel,lon=lonsel,method='nearest')
    qsel = Q['oceQnet'].sel(lat=latsel,lon=lonsel,method='nearest')
    HAdvsel = HAdv['HAdv'].sel(lat=latsel,lon=lonsel,method='nearest')
    H = zlayerToDepth(Zsel)
    ekman = zetasel / H * Thetasel[Zsel]
    heatflux = qsel / (H * rhoConst * Cp)
    Estimate = HAdvsel[Zsel]*-1 + ekman + heatflux
    return Estimate
def getThetaTendencyIntegral(ZselLowLLimit, Theta,latsel, lonsel):
    Integral = 0
    for d in range(0, ZselLowlimit+1):
        thickness = getZlayerThickness(d)
        Thetasel = (Theta['Theta'].sel(lat=latsel,lon=lonsel,method='nearest'))[d]
        ThetaTendencyAtLayerd = Thetasel[zsel].diff(dim='time') / 3600
        Integrand = ThetaTendencyAtLayerd*thickness
        Integral += Integrand
    return Integral
def getThetaTendencyEstimateIntegral(ZselLowLimit, HAdv, Theta,Zeta,Q, latsel,lonsel):
    Integral = 0
    zetasel = Zeta['Zeta'].sel(lat=latsel,lon=lonsel,method='nearest')
    qsel = Q['oceQnet'].sel(lat=latsel,lon=lonsel,method='nearest')
    heatflux = qsel / (H * rhoConst * Cp)
    for d in range(0, ZselLowlimit+1):
        Thetasel = (Theta['Theta'].sel(lat=latsel,lon=lonsel,method='nearest'))[d]
        HAdvsel = (HAdv['HAdv'].sel(lat=latsel,lon=lonsel,method='nearest'))[d]
        H = zlayerToDepth(Zsel)
        ekman = zetasel / H * Thetasel[Zsel]
        
        Estimate = HAdvsel[Zsel]*-1 + ekman + heatflux
    return Integral
```
